# 10 — Navidrome + SmartQueue validation
Runs in the **Jupyter container on node-serve-navidrome**.

```python
import requests
import json
```

```python
SERVING_URL = "http://fastapi_lgbm:8000"
NAVIDROME_URL = "http://host.docker.internal:4533"
API_BASE = f"{NAVIDROME_URL}/api"
```

```python
# Serving health check
r = requests.get(f"{SERVING_URL}/health", timeout=20)
print(r.status_code)
print(json.dumps(r.json(), indent=2))
```

```python
SESSION_ID = "test-session-001"
TEST_MEDIA_IDS = ["song-id-1", "song-id-2", "song-id-3"]
```

```python
# Start session via Navidrome
r = requests.post(
    f"{API_BASE}/smartqueue/session/start",
    json={"sessionId": SESSION_ID},
    timeout=20,
)
print(r.status_code, r.text)
```

```python
# Check active sessions
r = requests.get(f"{API_BASE}/smartqueue/session/active", timeout=20)
print(r.status_code)
print(json.dumps(r.json(), indent=2))
```

```python
# Rank request via Navidrome
rank_payload = {
    "sessionId": SESSION_ID,
    "ids": TEST_MEDIA_IDS,
}
r = requests.post(f"{API_BASE}/smartqueue/rank", json=rank_payload, timeout=30)
print(r.status_code)
if r.headers.get("content-type", "").startswith("application/json"):
    print(json.dumps(r.json(), indent=2))
else:
    print(r.text)
```

```python
# End session
r = requests.post(
    f"{API_BASE}/smartqueue/session/end",
    json={"sessionId": SESSION_ID},
    timeout=20,
)
print(r.status_code, r.text)
```

```python
# Confirm session count dropped
r = requests.get(f"{API_BASE}/smartqueue/session/active", timeout=20)
print(r.status_code)
print(json.dumps(r.json(), indent=2))
```
